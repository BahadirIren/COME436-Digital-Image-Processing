```python
class CameraThread(QThread):
    changePixmap = pyqtSignal(QImage)
    deletePixmap = pyqtSignal()
    
    def run(self):
        self.isThreadActive = True
        self.size = (640, 480)
        cap = cv2.VideoCapture(0)

        while self.isThreadActive:
            ret, self.frame = cap.read()

            if ret:
                rgbImage = cv2.cvtColor(self.frame, cv2.COLOR_BGR2RGB)
                h, w, ch = rgbImage.shape
                bytesPerLine = ch * w
                convertToQtFormat = QImage(
                    rgbImage.data, w, h, bytesPerLine, QImage.Format_RGB888)
                self.changePixmap.emit(convertToQtFormat)

        cap.release()
        self.deletePixmap.emit()

    def closeCamera(self):
        self.isThreadActive = False
        self.quit()

    def snapCamera(self):
        if self.isThreadActive:
            cv2.imwrite("images/snapped-image.png", self.frame)
```

```python
class MainWindow(QMainWindow):

    def __init__(self):
        super(MainWindow, self).__init__()

        self.ui = Ui_MainWindow()
        self.ui.setupUi(self)
        self.ui.retranslateUi(self)

        self.ui.imageLabel.setText("Hello,\n" + "Click Open button in Connect section to open camera\n" + 
                                    "or click Load Image button to load image from disk")

        # -------------- Connect Section --------------
        self.ui.openButton.clicked.connect(self.openCamera)
        self.ui.closeButton.clicked.connect(self.closeCamera)
        self.ui.snapButton.clicked.connect(self.snapCamera)

        # -------------- Image Processing Section --------------
        self.ui.loadImageButton.clicked.connect(self.loadImage)
        self.ui.cropAndSaveButton.clicked.connect(self.cropAndSave)
        self.ui.convertToGrayButton.clicked.connect(self.convertToGray)
        self.ui.convertToBlackAndWhiteButton.clicked.connect(self.convertToBlackAndWhite)
        self.ui.take2DFastFourierTransformButton.clicked.connect(self.take2DFastFourierTransform)
        self.ui.addGaussianNoiseButton.clicked.connect(self.addGaussianNoise)
        self.ui.addSaltAndPepperNoiseButton.clicked.connect(self.addSaltAndPepperNoise)
        self.ui.meanFilterButton.clicked.connect(self.meanFilter)
        self.ui.medianFilterButton.clicked.connect(self.medianFilter)
        self.ui.gaussianFilterButton.clicked.connect(self.gaussianFilter)
        self.ui.edgesButton.clicked.connect(self.edges)
        self.ui.cornersButton.clicked.connect(self.corners)
        

        # -------------- Extra Section --------------
        self.ui.helpButton.clicked.connect(self.help)
        self.ui.exitButton.clicked.connect(self.exit)

    

    # -------------- Connect Section --------------
    def deleteImage(self):
        self.ui.imageLabel.clear()
        self.ui.imageLabel.setText("Camera closed")

    @pyqtSlot(QImage)
    def setImage(self, image):
        self.ui.imageLabel.setPixmap(QPixmap.fromImage(image))

    def openCamera(self):
        self.thread = CameraThread(self)
        self.thread.changePixmap.connect(self.setImage)
        self.thread.deletePixmap.connect(self.deleteImage)
        self.thread.start()

        self.ui.snapButton.setEnabled(True)
        self.ui.openButton.setEnabled(False)
        self.ui.closeButton.setEnabled(True)
        self.setImageProcessingButtonsEnabled(False)

    def closeCamera(self):
        self.ui.imageLabel.setAlignment(Qt.AlignCenter)
        if isinstance(self.thread, CameraThread):
            self.thread.closeCamera()

        self.ui.snapButton.setEnabled(False)
        self.ui.openButton.setEnabled(True)
        self.ui.closeButton.setEnabled(False)


    def snapCamera(self):
        self.thread.snapCamera()

    # -------------- Image Processing Section --------------
    def loadImage(self):
        if isinstance(self.thread, CameraThread):
            self.closeCamera()

        self.fileName, _ = QFileDialog.getOpenFileName(self, 'Open file', '',
                                                        "Image files (*.jpg *.png *.jpeg)")
        if self.fileName:
            self.ui.imageLabel.setPixmap(QPixmap(self.fileName))

            self.ui.snapButton.setEnabled(False)
            self.ui.openButton.setEnabled(True)
            self.ui.closeButton.setEnabled(False)
            self.setImageProcessingButtonsEnabled(True)

    def cropAndSave(self):
        img = cv2.imread(self.fileName)
        
        img = cv2.resize(img, (self.ui.imageLabel.frameGeometry().width(), 
                                self.ui.imageLabel.frameGeometry().height()),
                                 interpolation= cv2.INTER_AREA)
        
        try: 
            width , height, _ = img.shape
        except:
            width, height = img.shape

        cropped_img = img[100: height-100 , 100: width-100]

        self.setImage(self.convertCv2ImageToQtFormat(cropped_img))
        cv2.imwrite("images/cropped-image.png", cropped_img)

    def convertToGray(self):
        img = cv2.imread(self.fileName)
        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        self.setImage(self.convertCv2ImageToQtFormat(gray_img))
        cv2.imwrite("images/gray-image.png", gray_img)

    def convertToBlackAndWhite(self):
        img = cv2.imread(self.fileName)
        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        _, blackAndWhiteImg = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)

        self.setImage(self.convertCv2ImageToQtFormat(blackAndWhiteImg))
        cv2.imwrite("images/black-and-white-image.png", blackAndWhiteImg)


    def take2DFastFourierTransform(self):
        img = cv2.imread(self.fileName, 0)
        dft = cv2.dft(np.float32(img),flags = cv2.DFT_COMPLEX_OUTPUT)
        dft_shift = np.fft.fftshift(dft)
        magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1]))

        self.setImage(self.convertCv2ImageToQtFormat(magnitude_spectrum))

    def convertCv2ImageToQtFormat(self, image):
        rgbImage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        h, w, ch = rgbImage.shape
        bytesPerLine = ch * w
        convertToQtFormat = QImage(
                    rgbImage.data, w, h, bytesPerLine, QImage.Format_RGB888)
        return convertToQtFormat

    def addGaussianNoise(self):
        img = cv2.imread(self.fileName)

        gauss = np.random.normal(0, 1, img.size)
        gauss = gauss.reshape(img.shape[0],img.shape[1],img.shape[2]).astype('uint8')
        imgGauss = cv2.add(img, gauss)
        
        self.setImage(self.convertCv2ImageToQtFormat(imgGauss))

        cv2.imwrite("images/gaussian-noise.png", imgGauss)


    def addSaltAndPepperNoise(self):
        img = cv2.imread(self.fileName, cv2.IMREAD_GRAYSCALE)
        
        try: 
            row , col, _ = img.shape
        except:
            row, col = img.shape
    
        # Pick a random number between 300 and 10000
        number_of_pixels = random.randint(300, 10000)
        for i in range(number_of_pixels):
    
            y_coord=random.randint(0, row - 1)
            x_coord=random.randint(0, col - 1)
            
            # Color that pixel to white
            img[y_coord][x_coord] = 255
            
        # Pick a random number between 300 and 10000
        number_of_pixels = random.randint(300 , 10000)
        for i in range(number_of_pixels):
    
            y_coord=random.randint(0, row - 1)
            x_coord=random.randint(0, col - 1)
            
            # Color that pixel to black
            img[y_coord][x_coord] = 0
            
        self.setImage(self.convertCv2ImageToQtFormat(img))
        cv2.imwrite("images/salt-and-pepper-noise-image.png", img)


    def meanFilter(self):
        img = cv2.imread(self.fileName)
        mean = cv2.blur(img,(5,5))

        self.setImage(self.convertCv2ImageToQtFormat(mean))

    def medianFilter(self):
        img = cv2.imread(self.fileName)
        median = cv2.medianBlur(img, 5)

        self.setImage(self.convertCv2ImageToQtFormat(median))

    def gaussianFilter(self):
        img = cv2.imread(self.fileName)
        gaussian = cv2.GaussianBlur(img, (5,5), 0)

        self.setImage(self.convertCv2ImageToQtFormat(gaussian))

    def edges(self):
        img = cv2.imread(self.fileName, 0)
        edges = cv2.Canny(img,100,200)

        self.setImage(self.convertCv2ImageToQtFormat(edges))


    def corners(self):
        img = cv2.imread(self.fileName)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        gray = np.float32(gray)
        dst = cv2.cornerHarris(gray,2,3,0.04)
        dst = cv2.dilate(dst,None)
        img[dst>0.01*dst.max()]=[0,0,255]

        self.setImage(self.convertCv2ImageToQtFormat(img))

    def setImageProcessingButtonsEnabled(self, isEnabled):
        self.ui.cropAndSaveButton.setEnabled(isEnabled)
        self.ui.convertToGrayButton.setEnabled(isEnabled)
        self.ui.convertToBlackAndWhiteButton.setEnabled(isEnabled)
        self.ui.take2DFastFourierTransformButton.setEnabled(isEnabled)
        self.ui.addGaussianNoiseButton.setEnabled(isEnabled)
        self.ui.addSaltAndPepperNoiseButton.setEnabled(isEnabled)
        self.ui.meanFilterButton.setEnabled(isEnabled)
        self.ui.medianFilterButton.setEnabled(isEnabled)
        self.ui.gaussianFilterButton.setEnabled(isEnabled)
        self.ui.edgesButton.setEnabled(isEnabled)
        self.ui.cornersButton.setEnabled(isEnabled)

    # -------------- Extra Section --------------
    def help(self):
        self.ui.imageLabel.setAlignment(Qt.AlignLeft)
        self.ui.imageLabel.setStyleSheet("padding :15px")
        self.ui.imageLabel.setText("- Connect Section:\n" +
                                     "\tOpen: Opens a camera\n" +
                                     "\tClose: Closes the camera\n"+
                                     "\tSnap: Take snapshot of live camera\n\n"+
                                     "- Image Processing Section:\n"+
                                     "\tLoad Image: Load images from disk\n"+
                                     "\tCrop and Save: Crops loaded image and saves to disk\n\n"+
                                     "\t- Image Analysis Section:\n"
                                     "\t\tConvert to Gray: Converts image to gray\n"+
                                     "\t\tConvert to Black and White: Converts image to black and white\n"+
                                     "\t\tTake 2-d fast Fourier Transform: Takes 2-d fast Fourier Transform of loaded image\n\n"+
                                     "\t\t- Add Noise Section:\n" +
                                     "\t\t\tGaussian: Adds gaussian noise to image\n"+
                                     "\t\t\tSalt and Pepper: Adds salt and pepper noise to image\n\n"+
                                     "\t\t\t- Filter Section:\n"+
                                     "\t\t\t\tMean: Applies mean filter to image\n" +
                                     "\t\t\t\tMedian: Applies median filter to image\n" +
                                     "\t\t\t\tGaussian: Applies gaussian filter to image\n"+
                                     "\tEdges: Finds edges of image\n"+
                                     "\tCorners: Finds corners of image\n\n"+
                                     "- Extra Section:\n"+
                                     "\tHelp: Opens help page\n"+
                                     "\tExit: Closes application\n")

    def exit(self):
        self.closeCamera()
        sys.exit()
```
